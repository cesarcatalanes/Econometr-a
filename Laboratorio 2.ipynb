## 4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diferencia = a - b

media_diferencia = np.mean(diferencia)
print(f"Media aritmética de la diferencia:{media_diferencia}")
```

## 5) Tengo valores de cordenadas (x, y) en las columnas del array
1.33 4.5
30.0 10.7
70.2 0.5
Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

coordenadas_nuevas = np.vstack([coordenadas, [37.1, -3.6]])

print(f"Dimensiones del nuevo array: {coordenadas_nuevas}")
```

## 6) Copie el array del ejercicio anterior. Traspóngalo (Agregue ahora dos nuevos pares de coordenadas: (10.8, 3.0) y (35.8, 12.0)

## 1. ¿Cuántos Municipios tienen más de 100,000 habitantes?


```python
import pandas as pd

df = pd.read_csv("NucleosPoblacion.csv")
municipios_mas_100k = df[df['Poblacion'] > 100000]
cantidad_municipios = len(municipios_mas_100k)
print(f"Municipios con más de 100,000 habitantes: {cantidad_municipios}")
```

## 2. Gráfica de barras de la población de cada ciudad ordenada de menor a mayor
a. ¿Cuál es la segunda ciudad más poblada?
b. ¿Qué posición ocupa Granada en el ranking?

```python
import matplotlib.pyplot as plt

df_ordenado = df.sort_values(by='Poblacion')

plt.figure(figsize=(10, 6))
plt.bar(df_ordenado['Municipio'], df_ordenado['Poblacion'])
plt.xticks(rotation=90)
plt.title('Población de cada ciudad (ordenada de menor a mayor)')
plt.xlabel('Municipio')
plt.ylabel('Población')
plt.show()

segunda_ciudad = df.sort_values(by='Poblacion', ascending=False).iloc[1]['Municipio']
print(f"La segunda ciudad más poblada es: {segunda_ciudad}")

ranking_granada = df.sort_values(by='Poblacion', ascending=False).reset_index()
posicion_granada = ranking_granada[ranking_granada['Municipio'] == 'Granada'].index[0] + 1
print(f"Granada ocupa la posición: {posicion_granada}")
```

## 3. ¿Cuántos municipios de la provincia de Tarragona tienen más de 32,000 habitantes?


```python
tarragona_32000 = df[(df['Provincia'] == 'Tarragona') & (df['Poblacion'] > 32000)]
cantidad_tarragona = len(tarragona_32000)
print(f"Municipios en Tarragona con más de 32,000 habitantes: {cantidad_tarragona}")
```

## 4. ¿Cuál es el municipio situado más al Norte?


```python
municipio_norte = df.loc[df['Y'].idxmax()]
print(f"Municipio más al Norte: {municipio_norte['Municipio']}, Provincia: {municipio_norte['Provincia']}, Población: {municipio_norte['Poblacion']}")
```

## 5. Estadísticas de los municipios de Granada


```python
granada_municipios = df[df['Provincia'] == 'Granada']

media = granada_municipios['Poblacion'].mean()
mediana = granada_municipios['Poblacion'].median()
desviacion_std = granada_municipios['Poblacion'].std()
valor_max = granada_municipios['Poblacion'].max()
valor_min = granada_municipios['Poblacion'].min()

print(f"Media: {media}, Mediana: {mediana}, Desviación estándar: {desviacion_std}, Máximo: {valor_max}, Mínimo: {valor_min}")
```

## 6. Histograma de la población por provincia


```python
provincias = df['Provincia'].unique()
for provincia in provincias:
    plt.figure(figsize=(8, 5))
    plt.hist(df[df['Provincia'] == provincia]['Poblacion'], bins=10, alpha=0.7, label=provincia)
    plt.title(f"Histograma de población - Provincia: {provincia}")
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.show()
```

## 7. Selección aleatoria de 50 municipios

```python
sample_50 = df.sample(50, random_state=42)

promedio_poblacion = sample_50['Poblacion'].mean()
desviacion_muestra = sample_50['Poblacion'].std()

print(f"Promedio de población: {promedio_poblacion}, Desviación estándar: {desviacion_muestra}")
```

## 8. Municipios más cercano y más lejano a Madrid


```python
madrid_coords = df[df['Municipio'] == 'Madrid'][['X', 'Y']].iloc[0]

df['Distancia_a_Madrid'] = np.sqrt((df['X'] - madrid_coords['X'])**2 + (df['Y'] - madrid_coords['Y'])**2)

df_sin_madrid = df[df['Municipio'] != 'Madrid']
mas_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]
mas_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]

print(f"Más cercano a Madrid: {mas_cercano['Municipio']}")
print(f"Más lejano a Madrid: {mas_lejano['Municipio']}")
```

## 9. Scatter plot de las coordenadas de Madrid y Barcelona


```python
madrid = df[df['Provincia'] == 'Madrid']
barcelona = df[df['Provincia'] == 'Barcelona']

plt.figure(figsize=(8, 6))
plt.scatter(madrid['X'], madrid['Y'], color='blue', label='Madrid')
plt.scatter(barcelona['X'], barcelona['Y'], color='red', label='Barcelona')
plt.title('Municipios de Madrid y Barcelona')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()
plt.show()
```
